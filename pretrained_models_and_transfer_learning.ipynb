## 1. ImageNet: A Pioneering Vision for Computers
It all began in 1985. George A. Miller and his team at Princeton University started working on WordNet, a lexical database for the English language. As a cross between a dictionary and a thesaurus, it enabled applications in Natural Language Processing (NLP).

Fast forward 21 years and data scientist Fei-Fei Li came up with the idea of ImageNet at the University of Illinois Urbana-Champaign. At the time, most AI researchers thought algorithms were more important than the data itself. However, Li was convinced that vast amounts of real-world data would make algorithms more accurate. By then, WordNet was mature, and the version 3.0 had recently been released. When Li met WordNet researcher Christiane Fellbaum from Princeton University, she decided to use the wordbase and hierarchy of WordNet for her ambitious image database. Her aim? Support visual object recognition software research.

#### Collective Contributions With Spectacular Results

In July 2008, ImageNet had zero images. By December, it had categorized three million images across 6,000+ synsets. In April 2010, there were more than 11 million images in 15,000+ synsets. Such results would have been inconceivable for a handful of researchers. They were made possible through crowdsourcing on Amazonâ€™s Mechanical Turk platform.

In 2010, the first ever ImageNet Large Scale Visual Recognition Challenge (ILSVRC) was organized. Software programs competed to correctly classify and detect objects and scenes.

Since being launched, ImageNet has given researchers a common set of images to benchmark their models and algorithms. In turn, this has driven research in machine learning and deep neural networks, making it easier to classify images and complete other tasks associated with computer vision. The data is available for free to researchers for non-commercial use.

## 2. Pretrained models

Using pretrained models in deep learning offers several advantages over creating models from scratch, as highlighted below:

### **Advantages of Pretrained Models**

1. **Reduced Training Time**:
    - Pretrained models are already trained on large datasets, so they have learned meaningful features. Fine-tuning them for a specific task requires significantly less time compared to training from scratch.
    - For example, training a model from scratch can take months, while fine-tuning a pretrained model can be accomplished in weeks or even days.
2. **Lower Computational Costs**:
    - Training deep learning models from scratch requires massive computational resources and expensive hardware. Pretrained models reduce these requirements by leveraging precomputed weights and features.
3. **Better Performance with Limited Data**:
    - Pretrained models are particularly useful when labeled data is scarce. They transfer knowledge from large datasets to smaller, domain-specific datasets, improving performance even with limited training data.
4. **Improved Accuracy**:
    - Pretrained models often achieve state-of-the-art performance because they start with weights optimized on large, diverse datasets like ImageNet. This provides a strong foundation for fine-tuning on specialized tasks.
5. **Ease of Use**:
    - Developers can use pretrained models as-is or adapt them for specific use cases through transfer learning. This simplifies the development process and makes machine learning more accessible to non-experts.
6. **Knowledge Transfer**:
    - Pretrained models act as a form of transfer learning, where knowledge gained from one domain (source) is applied to another domain (target). This is especially effective in tasks like natural language processing, computer vision, and speech recognition.
7. **Cost Efficiency**:
    - By reducing the need for extensive data collection, annotation, and computational resources, pretrained models save significant costs in AI development.
### **Applications of Pretrained Models**

- **Natural Language Processing (NLP)**: Used in translation, chatbots, and large language models like GPT.
- **Computer Vision**: Applied in object detection, image segmentation, and classification.
- **Healthcare**: Accelerates drug discovery and medical image analysis.
- **Cybersecurity**: Detects anomalies and phishing attacks.
- **Creative Workflows**: Powers AI art tools and generative design systems.


### **Challenges of Training From Scratch**

- Requires vast amounts of labeled data.
- Demands high computational power.
- Involves complex optimization processes that are time-consuming and prone to overfitting when data is limited.

## 4. How CNN see?




Deep learning networks often work like a "black box." They can make good predictions, but it's hard to understand why they made a particular prediction.

However, with convolutional neural networks (CNNs), which are used for images, we have a little more insight. CNNs are built to understand images in a way that preserves the relationships between parts of the image e.g., spatial relationships. (like how edges and textures are arranged).

We can actually look at parts of the CNN to see what it has learned. We can visualize the "filters" and see what kind of features the CNN is designed to find (like edges, colors, or textures). Also, when we give the CNN an image, we can look at the "activation maps" (think of them as notes the CNN takes) to see exactly what features the CNN detected in that specific image.

We need a model to visualize.
Instead of fitting a model from scratch, we are using a pre-trained image classification model VGG16.

```python
import tensorflow as tf
from keras.applications.vgg16 import VGG16, preprocess_input

model = VGG16()
```

```python
print(model.summary())
```

The model summary printed summarizes the output shape of each layer, e.g. the shape of the resulting feature maps. It does not give any idea of the shape of the filters (weights) in the network, only the total number of weights per layer.

We can access all of the layers of the model via the model.layers property.

```python
# summarize filter shapes
for layer in model.layers:
	# check for convolutional layer
    if 'conv' not in layer.name:
        continue
# get filter weights
    filters, biases = layer.get_weights()
    print(layer.name, filters.shape)
```

### 4.1 Visualizing Filters
We can see that all convolutional layers use 3Ã—3 filters, which are small and perhaps easy to interpret.

An architectural concern with a convolutional neural network is that the depth of a filter must match the depth of the input for the filter (e.g. the number of channels).|

First convolution layer has 64 filters, with 3x3 size for each channel of Red, Green and Blue. <br>
Visualising filters of first convolution layer..

```python
filters, biases = model.layers[1].get_weights()
print("1st Conv Layer:", filters.shape)
```

```python
# visualising 8 out of 64 filters of the first convolutional layer

import matplotlib.pyplot as plt
f_max, f_min = filters.max(), filters.min()

# applying min - max normalization to scale values between 0-1
filter = (filters - f_min) / (f_max - f_min)
plt.figure(1, figsize=(8, 8))
n_filters, ix = 8, 1
for i in range(n_filters):
    # get the filter
    f = filter[:,:,:,i]

    # plot each channel separately
    for j in range(filters.shape[2]):
        # Subplots
        ax = plt.subplot(n_filters, filters.shape[2], ix)
        ax.set_xticks([])
        ax.set_yticks([])
        # plot filter channel
        plt.imshow(f[:,:,j], cmap='gray')
        ix+=1

plt.show()
```

### 4.2 Visualizing Feature maps
The activation maps, called feature maps, capture the result of applying the filters to input, such as the input image or another feature map.

The idea of visualizing a feature map for a specific input image would be to understand what features of the input are detected or preserved in the feature maps. The expectation would be that the feature maps close to the input detect small or fine-grained detail, whereas feature maps close to the output of the model capture more general features.

In order to explore the visualization of feature maps, we need input for the VGG16 model that can be used to create activations.

We will use a simple photograph.

```python
import skimage.io as io
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = io.imread("https://six5six.in/cdn/shop/files/INDIA-JerseyLarge_1024x.jpg?v=1697440965")
plt.imshow(img); plt.xticks([]); plt.yticks([]); plt.show()
```

Enumerating all convolution layers in the model and print the output size or feature map size for each convolutional layer as well as the layer index in the model.

```python
# understanding the input layer shape and output shapes of each convolution layer
for i in range(len(model.layers)):
    layer = model.layers[i]
    if ('conv' in layer.name) or ('input' in layer.name):
        print(f"{i}.", layer.name, layer.output.shape)
```

we need to load the bird image with the size expected by the model, in this case, 224Ã—224, after  seeiing the input

```python
# resizing image
img = cv2.resize(img, (224, 224))
img = img.reshape((1,)+ img.shape) # converting into batch of 1 image
img = preprocess_input(img) # preprocessing input as per VGG16
print(img.shape)
```

```python
# this is how a preprocessed image looked like
plt.imshow(img.squeeze()); plt.xticks([]); plt.yticks([]); plt.show()
```

```python
# redefine model to output right after the first hidden layer
from keras.models import Model
model_redf = Model(inputs=model.inputs, outputs=model.layers[1].output)
```

```python
model_redf.summary()
```

```python
feature_maps1 = model_redf.predict(img)
```

```python
feature_maps1.shape
```

```python
# plotting all 64 features into 8x8 squares
def plot_feature_maps(feature_maps, figsize=(12,12)):
    import matplotlib.pyplot as plt
    sq = 8
    ix = 1
    plt.figure(figsize=figsize)
    for _ in range(sq):
        for _ in range(sq):
            ax = plt.subplot(sq, sq, ix)
            ax.set_xticks([])
            ax.set_yticks([])
            plt.imshow(feature_maps[:, :, :, ix-1].squeeze(), cmap='gray')
            ix += 1
    plt.show()

plot_feature_maps(feature_maps1)
```

This is an interesting result and generally matches our expectation. We could update the example to plot the feature maps from the output of other specific convolutional layers.

There are five main blocks in the image (e.g. `block1`, `block2`, etc.) that end in a pooling layer. The layer indexes of the last convolutional layer in each block are `[2, 5, 9, 13, 17]`.

We can define a new model that has multiple outputs, one feature map output for each of the last convolutional layer in each block before pooling.

```python
# redefine model to output right after the first hidden layer of fully connected layer

ixs = [2, 5, 9, 13, 17]
outputs = [model.layers[i].output for i in ixs]
model_redf2 = Model(inputs=model.inputs, outputs=outputs)
model_redf2.summary()
```

Making a prediction with this new model will result in a list of feature maps.

Also, we know that the number of feature maps (e.g. depth or number of channels) in deeper layers is much more than $64$, such as $256$ or $512$. We can cap the number of feature maps visualized at $64$ for consistency.

```python
feature_maps2 = model_redf2.predict(img)
```

```python
for fmap in feature_maps2:
    plot_feature_maps(fmap)
    print("-"*100)
```

```python
from keras.applications.vgg16 import decode_predictions
print("Predictions: ",decode_predictions(model.predict(img), top=3))
```

This behavior is expected due to how convolutional neural networks (CNNs) process images layer by layer.

#### Why is the first image (early layers) brighter?
- **Low-Level Features**: The initial convolutional layers of VGG16 detect simple patterns like edges, textures, and corners.

- **Higher Activation Values**: Since these patterns are present across the image, many neurons in the feature maps are activated, resulting in bright, high-contrast outputs.

#### Why is the second image (deeper layers) darker?
- **High-Level Features**: As we move deeper into the network, the layers capture more complex and abstract features like object parts and shapes.

- **Sparse Activations**: Only specific regions (like eyes, nose, mouth) remain important for classification, while most neurons suppress irrelevant background information.

- **ReLU & Pooling Effects**: ReLU activation zeros out many values, and max pooling keeps only the strongest signals, leading to a sparsity effectâ€”darker visualizations.

In short, early layers highlight broad patterns across the image, while later layers focus only on the most discriminative features, making the activations sparser and darker.

## 5. Transfer Learning
It is a research problem in machine learning that focuses on storing knowledge gained while solving one problem and applying it to a different but related problem.

It is a machine learning technique where a pre-trained model (trained on a large dataset like ImageNet) is reused for a new but related task. Instead of training a model from scratch, we take a model that has already learned useful features and fine-tune it for our specific dataset.

In the context of deep learning, transfer learning typically involves:

- Taking a pre-trained model (e.g., VGG16, ResNet, or BERT) that was trained on a large dataset (e.g., ImageNet for images).

- Fine-tuning or adapting the model to a new, often smaller, dataset or task.


### 5.1 Why Do We Need Transfer Learning?
1. **Saves Time and Computational Resources** :
    
    - Training deep learning models from scratch requires significant computational power and time, especially for large datasets.

    - Transfer learning allows you to start with a pre-trained model, which has already learned useful features, reducing the training time and resource requirements.

2. **Works Well with Small Datasets** :
    
    - Deep learning models typically require large amounts of labeled data to perform well. However, in many domains (e.g., medical imaging, satellite imagery), labeled data is scarce or expensive to obtain.

    - Transfer learning enables you to use a pre-trained model that has already learned general features from a large dataset (e.g., ImageNet) and fine-tune it on a smaller dataset for a specific task.

3. **Improves Performance** :

    - Pre-trained models have already learned useful features (e.g., edges, textures, shapes for images or word embeddings for text) from large datasets.

    - By leveraging these features, transfer learning often leads to better performance on the target task compared to training a model from scratch, especially when the target dataset is small.

4. **Avoids Overfitting** :

    - When training on small datasets, models are prone to overfitting (memorizing the training data instead of generalizing).

    - Transfer learning helps mitigate this by starting with a model that has already learned general patterns, reducing the risk of overfitting.

5. **Cross domain adaption** :

    - Transfer learning allows knowledge to be transferred across different but related domains. For example:





### 5.2 **Real world Example - DeepFace**
**DeepFace**, developed by **Facebook (now Meta)**, is a **real-world example of transfer learning** in action. It is a deep learning-based facial recognition system used for tasks like tagging friends in photos, detecting unauthorized use of profile pictures, and enhancing security through facial recognition login.

#### How DeepFace Uses Transfer Learning

##### 1. Pre-trained Model on a Large Dataset
   - DeepFace leverages a **pre-trained deep convolutional neural network (CNN) i.e., VGGFace**  trained on a massive dataset of labeled faces.
   - The model is initially trained on a general-purpose face dataset (e.g., **Labeled Faces in the Wild (LFW)** or **Facebook's internal dataset**) to learn **general facial features** such as edges, textures, and facial structures.

##### 2. Feature Extraction
   - The pre-trained model learns to extract **high-level features** from faces, such as:
     - Distance between eyes.
     - Shape of the nose.
     - Jawline structure.
   - These features are **generic** and can be applied to any face recognition task.

##### 3. Fine-Tuning for Specific Tasks
   - DeepFace fine-tunes the pre-trained model on **Facebook's specific dataset** of user-uploaded photos.
   - During fine-tuning:
     - The model adapts to the specific characteristics of Facebook's data (e.g., lighting conditions, angles, and resolutions commonly found in user photos).
     - The final layers of the model are adjusted to improve accuracy for Facebook's use cases (e.g., tagging friends or detecting fake accounts).

##### 4. Transfer of Knowledge
   - The knowledge (learned features) from the pre-trained model is **transferred** to the new task of recognizing faces in Facebook's environment.
   - This avoids the need to train a model from scratch, which would require:
     - A massive dataset of labeled faces.
     - Significant computational resources and time.

#### How Facenet is Used in DeepFace

- Face Detection: The input images are processed to detect faces using a face detection backend (e.g., OpenCV, retinaface, mtcnn, yolov8).

- Feature Extraction: The detected faces are passed through the Facenet model to generate face embeddings (128-dimensional vectors representing the facial features).

- Similarity Calculation: The cosine similarity metric is used to compare the face embeddings of two images. If the distance between the embeddings is below a certain threshold, the faces are considered a match.

- Verification Result: The result ('verified': True or 'verified': False) is returned, along with additional details like the distance, threshold, and facial areas.

#### What DeepFace.verify Does?

DeepFace.verify is a complete pipeline that integrates multiple steps to perform face verification. Here's how it works:

1. **Face Detection:**
    
    Uses a face detection backend (e.g., OpenCV, Dlib, or MTCNN) to locate faces in the input images.

    **Example**: Detects the bounding box around a face in img1_path and img2_path.

2. **Face Alignment (Optional):**

    Aligns the detected faces to a standard pose (e.g., frontal view) to improve accuracy.

    Example: Rotates the face so that the eyes are horizontally aligned.

3. **Face Embedding Generation:**

    Uses a pre-trained model (e.g., Facenet, VGG-Face) to generate face embeddings.

    Example: Passes the aligned face through Facenet to generate a 128-dimensional embedding.

4. **Similarity Calculation:**

    Compares the embeddings of the two faces using a similarity metric (e.g., cosine similarity, Euclidean distance).

    Example: Calculates the cosine similarity between the embeddings of img1_path and img2_path.

5. **Verification:**

    Compares the similarity score to a predefined threshold to determine if the faces belong to the same person.

    Example: If the cosine similarity is above the threshold (e.g., 0.40 for Facenet), the faces are verified as the same person.

**Que. Why Facenet Alone is Not Enough**
- Facenet is just the embedding model. It cannot:
  - Detect faces in an image.
  - Align faces to a standard pose.
  - Compare embeddings or apply a threshold for verification.
- These additional steps are essential for a complete face verification pipeline, which is why `DeepFace.verify` is needed.

### 5.3 How Transfer learning works

We have a model train on Cat and Dog image classification, imagine we have a second dataset that shows pictures of cows and horses. Cows and horses are significantly different from cats.<br>
Nevertheless, they are all mammals, they have four feet, and a similar shape. As a solution, we can take the layers that describe the shape of the object we want to detect, whether it is a dog, a cow, or a horse, and freeze them. Freezing means we cut them out of our predictive function $f$ put them in our predictive function $g$, and train the function $g$, without training our frozen layers:

We can see the layers on the right side are green, which indicates that we have to train them while creating our predictive function for the source domain. On the other hand, the predictive $g$ is created using the already existing, frozen layer from the source domain. The frozen layers stay untouched during the training process.

### 5.4 How Does Transfer Learning Work?
Transfer learning is a machine learning technique where a pre-trained model (trained on a large dataset) is fine-tuned or used as a feature extractor for a new task. Instead of training a model from scratch, transfer learning leverages knowledge from a source task to improve performance on a target task.

#### Two Main Methods of Transfer Learning
1. Feature Extraction (Using Pre-trained Model as a Fixed Feature Extractor)
2. Fine-tuning (Retraining Some or All Layers)

#### **1. Feature Extraction (Frozen Layers)**
* A pre-trained model (like ResNet, VGG, or BERT) is used as a fixed feature extractor.
* The lower layers of the model (which have learned generic features like edges, shapes, and textures) remain frozen.
* Only the final layers (classifier or regressor) are replaced and trained on the new dataset.

**e.g.**, We have a small dataset of X-ray images and want to classify them as normal or pneumonia. Since X-ray datasets are small, training a model from scratch is not feasible.<br>
Instead, we use a pre-trained CNN (like VGG16) and replace its classifier. The early layers of VGG16 already detect edges & textures, which work well for X-ray images, so we don't want to retrain all the layers.

We'll pass our images through VGG16's convolutional layers to extract convolved visual features, which will output a Feature Stack of the detected visual features. From here, it's easy to flatten the 3-Dimensional feature stack into a NumPy array - ready to feed into our new fully connected layers.


#### **2. Fine-Tuning (Unfreezing Some or All Layers)**

* Instead of keeping all layers frozen, some deeper layers are unfrozen and retrained on the new dataset.
* This allows the model to adjust its feature extraction to the specific characteristics of the new data.
* The first few layers remain frozen since they capture generic features (edges, textures), while deeper layers are fine-tuned.

The frozen pre-trained layers will convolve visual features as usual. The non-frozen (i.e., the 'trainable') pre-trained layers will be trained on our custom dataset and update according to the Fully-Connected layer's predictions.

**e.g.**, We are building an autonomous vehicle that detects traffic signs. The available dataset is moderate-sized (10,000+ images), and signs may have different angles/brightness than those in ImageNet.<br>
We fine-tune the deeper layers of ResNet to adapt to new patterns. We only unfreeze the last few layers, so the model learns traffic signs while retaining general features.

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
from keras.applications.vgg16 import VGG16
```

```python
# using VGG16 pretrained model
conv_base = VGG16(weights='imagenet', include_top = False, input_shape=(150,150,3))
# include_top=False: Removes the final classification layers of VGG16, Useful for transfer learning on custom datasets
```

```python
conv_base.summary()
```

```python
# freezing the convolutional layers
conv_base.trainable = False
```

No flatten and fully connected layer here.

Creating our custom fully connected layer network

```python
model1 = Sequential()

model1.add(conv_base)
model1.add(Flatten())
model1.add(Dense(256,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
# generators for feeding images in batch to model

train_ds = keras.utils.image_dataset_from_directory(
    directory = 'data/cat_n_dog/train',
    labels='inferred',
    label_mode = 'int',
    batch_size=32,
    image_size=(150,150)
)

validation_ds = keras.utils.image_dataset_from_directory(
    directory = 'data/cat_n_dog/test',
    labels='inferred',
    label_mode = 'int',
    batch_size=32,
    image_size=(150,150)
)
```

```python
def process(image, label):
    image = tensorflow.cast(image/255. , tensorflow.float32)
    return image, label

train_ds = train_ds.map(process)
validation_ds = validation_ds.map(process)
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history = model1.fit(train_ds, epochs = 10, validation_data = validation_ds)
```

#### 5.5.2 **feature extraction transfer learning(with data augmentation)**

__Data augmentation__ is the process of increasing the amount and diversity of data. We do not collect new data, rather we transform the already present data. Data means, image data here.

**How it works**

- Data augmentation makes small changes to the original data, such as cropping, rotating, Shearing, Zooming, brightness level and flipping.
- It can be used in training, testing, or both.
- It can help prevent overfitting, which is when a model performs well on training data but not on unseen data.

**Benefits**

- It can improve model optimization and generalizability.
- It can prevent overfitting by exposing models to a wider range of scenarios.
- It can help models perform better on low-resources.
- It's cost-effective because it doesn't require collecting and annotating a large, diverse datase

```python
model2 = Sequential()

model2.add(conv_base)
model2.add(Flatten())
model2.add(Dense(256,activation='relu'))
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
# data augmentation generator
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img

batch_size = 32

train_data_gen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_data_gen = ImageDataGenerator(rescale=1./255)

train_generator = train_data_gen.flow_from_directory(
    'data/cat_n_dog/train', target_size=(150, 150), batch_size=batch_size, class_mode='binary')

test_generator = test_data_gen.flow_from_directory(
    'data/cat_n_dog/test', target_size=(150, 150), batch_size=batch_size, class_mode='binary')
```

```python
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history2 = model2.fit(train_generator, epochs=10, validation_data=test_generator)
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
# Create the figure and subplots
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot for loss
ax[0].plot(history2.history['loss'], label='train')
ax[0].plot(history2.history['val_loss'], label='val')
ax[0].set(title='Training and Validation Loss', xlabel='Epochs', ylabel='Loss')
ax[0].legend()

# Plot for accuracy
ax[1].plot(history2.history['accuracy'], label='train')
ax[1].plot(history2.history['val_accuracy'], label='val')
ax[1].set(title='Training and Validation Accuracy', xlabel='Epochs', ylabel='Accuracy')
ax[1].legend()

# Display the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
